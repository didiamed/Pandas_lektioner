##### Correlation caveats

<span style="color:blue"> 1. Varningar för korrelation </span>

Även om korrelation är ett användbart sätt att kvantifiera relationer finns det några varningar att tänka på.

<span style="color:blue"> 2. Icke-linjära relationer </span>

Titta på den här datan. Det finns tydligt en relation mellan x och y, men när vi beräknar korrelationen får vi 0-komma-18.

<span style="color:blue"> 3. Icke-linjära relationer </span>

Det beror på att relationen mellan de två variablerna är en kvadratisk relation, inte en linjär relation. Korrelationskoefficienten mäter styrkan hos linjära relationer och endast linjära relationer.

<span style="color:blue"> 4. Korrelation tar bara hänsyn till linjära relationer </span>

Precis som alla sammanfattande statistik bör korrelation inte användas blindt, och du bör alltid visualisera din data när det är möjligt.

<span style="color:blue"> 5. Data om sömn hos däggdjur </span>

Låt oss återvända till datan om sömn hos däggdjur.

<span style="color:blue"> 6. Kroppsvikt mot vaken tid </span>

Här är en spridningsplot av varje däggdjurs kroppsvikt mot den tid de är vakna varje dag. Relationen mellan dessa variabler är definitivt inte linjär. Korrelationen mellan kroppsvikt och vaken tid är endast cirka 0-komma-3, vilket är en svag linjär relation.

<span style="color:blue"> 7. Fördelning av kroppsvikt </span>

Om vi tittar närmare på fördelningen av kroppsvikt ser vi att den är starkt snedvriden. Det finns många lägre vikter och några vikter som är mycket högre än resten.

<span style="color:blue"> 8. Logaritmisk transformation </span>

När data är snedvriden på det här sättet kan vi tillämpa en logaritmisk transformation. Vi skapar en ny kolumn som heter log_bodywt som innehåller logaritmen av varje kroppsvikt. Vi kan göra detta med hjälp av np-dot-log. Om vi plottar logaritmen av kroppsvikt mot vaken tid ser relationen mycket mer linjär ut än mellan vanlig kroppsvikt och vaken tid. Korrelationen mellan logaritmen av kroppsvikt och vaken tid är cirka 0-komma-57, vilket är mycket högre än de 0-komma-3 vi hade tidigare.

<span style="color:blue"> 9. Andra transformationer </span>

Förutom logaritmisk transformation finns det många andra transformationer som kan användas för att göra en relation mer linjär, som att ta kvadratroten eller den reciproka av en variabel. Valet av transformation beror på datan och hur snedvriden den är. Dessa kan tillämpas i olika kombinationer på x och y, till exempel kan du tillämpa en logaritmisk transformation på både x och y, eller en kvadratrots-transformation på x och en reciprokk-transformation på y.

<span style="color:blue"> 10. Varför använda en transformation? </span>

Så varför använda en transformation? Vissa statistiska metoder bygger på att variabler har en linjär relation, som att beräkna en korrelationskoefficient. Linjär regression är en annan statistisk teknik som kräver att variabler är relaterade på ett linjärt sätt, vilket du kan lära dig allt om i den här kursen.

<span style="color:blue"> 11. Korrelation innebär inte kausalitet </span>

Låt oss prata om ännu en viktig begränsning med korrelation som du kanske har hört talas om tidigare: korrelation innebär inte kausalitet. Det betyder att om x och y är korrelerade betyder det inte nödvändigtvis att x orsakar y. Till exempel, här är en scatterplot av per capita-konsumtionen av margarin i USA varje år och skilsmässoraten i delstaten Maine. Korrelationen mellan dessa två variabler är 0-komma-99, vilket är nästan perfekt. Men det betyder inte att konsumera mer margarin kommer att orsaka fler skilsmässor. Den här typen av korrelation kallas ofta en falsk korrelation.

<span style="color:blue"> 12. 1 Sammanblandning </span>

En fenomen som kallas sammanblandning kan leda till falska korrelationer. Säg att vi vill veta om att dricka kaffe orsakar lungcancer. När vi tittar på data ser vi att kaffeintag och lungcancer är korrelerade, vilket kan få oss att tro att att dricka mer kaffe ger dig lungcancer.

<span style="color:blue"> 12. 2 Sammanblandning </span>

Men det finns en tredje, dold variabel som spelar roll, nämligen rökning.

<span style="color:blue"> 12. 3 Sammanblandning </span>

Det är känt att rökning är förknippad med kaffeintag.

<span style="color:blue"> 12. 4 Sammanblandning </span>

Det är också känt att rökning orsakar lungcancer.

<span style="color:blue"> 12. 5 Sammanblandning </span>

I verkligheten visar det sig att kaffe inte orsakar lungcancer utan bara är förknippat med det, men det verkade vara kausalt på grund av den tredje variabeln, rökning. Denna tredje variabel kallas en sammanblandare eller dold variabel. Det innebär att den intressanta relationen mellan kaffe och lungcancer är en falsk korrelation. Ett annat exempel på detta är relationen mellan helgdagar och försäljning inom detaljhandeln. Även om det kan vara så att folk köper mer runt helgdagar som ett sätt att fira, är det svårt att avgöra hur mycket av den ökade försäljningen beror på helgdagar och hur mycket som beror på speciella erbjudanden och kampanjer som ofta pågår kring helgdagar. Här förvränger speciella erbjudanden relationen mellan helgdagar och försäljning.

##### Design av experiment

<span style="color:blue">1. Design av experiment</span>

Ofta skapas data som ett resultat av en studie som syftar till att besvara en specifik fråga. Men data måste analyseras och tolkas olika beroende på hur datan genererades och hur studien var utformad.

<span style="color:blue">2. Ordlista </span>

Experiment syftar generellt till att besvara en fråga i form av "Vad är effekten av behandlingen på responsen?" I denna kontext avser behandling den förklarande eller oberoende variabeln, och respons avser responsen eller den beroende variabeln. Till exempel, vad är effekten av en annons på antalet köpta produkter? I detta fall är behandlingen en annons, och responsen är antalet köpta produkter.

<span style="color:blue">3. Kontrollerade experiment </span>

I ett kontrollerat experiment tilldelas deltagarna slumpmässigt antingen till behandlingsgruppen eller kontrollgruppen, där behandlingsgruppen får behandlingen och kontrollgruppen inte får det. Ett bra exempel på detta är ett A/B-test. I vårt exempel kommer behandlingsgruppen att se en annons, medan kontrollgruppen inte kommer att göra det. Förutom denna skillnad bör grupperna vara jämförbara så att vi kan avgöra om att se en annons får människor att köpa mer. Om grupperna inte är jämförbara kan detta leda till förvirring eller bias. Om genomsnittsåldern för deltagarna i behandlingsgruppen är 25 år och genomsnittsåldern för deltagarna i kontrollgruppen är 50 år, kan ålder vara en potentiell förvirringsfaktor om yngre personer är mer benägna att köpa mer, och detta kommer att göra experimentet biased mot behandlingen.

<span style="color:blue">4. Guldstandarden för experiment kommer att använda... </span>

Guldstandarden, eller det idealiska experimentet, kommer att eliminera så mycket bias som möjligt genom att använda vissa verktyg. Det första verktyget som hjälper till att eliminera bias i kontrollerade experiment är att använda en randomiserad kontrollerad studie. I en randomiserad kontrollerad studie tilldelas deltagarna slumpmässigt till behandlingsgruppen eller kontrollgruppen och deras tilldelning baseras inte på något annat än chans. En sådan slumpmässig tilldelning hjälper till att säkerställa att grupperna är jämförbara. Det andra sättet är att använda en placebo, vilket är något som liknar behandlingen men inte har någon effekt. På det sättet vet deltagarna inte om de tillhör behandlingsgruppen eller kontrollgruppen. Detta säkerställer att behandlingens effekt beror på behandlingen själv, inte idén om att få behandlingen. Detta är vanligt förekommande i kliniska prövningar som testar effektiviteten av ett läkemedel. Kontrollgruppen får fortfarande en tablett, men det är en sockerpiller som har minimala effekter på responsen.

<span style="color:blue">5. Guldstandarden för experiment använder också... </span>

I ett dubbelblindstudie vet inte heller personen som administrerar behandlingen eller utför experimentet om de administrerar den faktiska behandlingen eller placebon. Detta skyddar mot bias i responsen samt analysen av resultaten. Dessa olika verktyg kommer alla ned till samma princip: om det finns färre möjligheter för bias att smyga sig in i ditt experiment, desto mer tillförlitligt kan du dra slutsatsen om huruvida behandlingen påverkar responsen.


<span style="color:blue">6. Observationsstudier </span>

Den andra typen av studie vi ska diskutera är observationsstudien. I en observationsstudie tilldelas deltagarna inte slumpmässigt till grupper. Istället tilldelar deltagarna sig själva, vanligtvis baserat på befintliga egenskaper. Detta är användbart för att besvara frågor som inte lämpar sig för en kontrollerad experiment. Om du vill studera effekten av rökning på cancer kan du inte tvinga människor att börja röka. På samma sätt kan du inte tvinga människor att ha viss tidigare köpbeteende om du vill studera hur tidigare köpbeteende påverkar om någon kommer att köpa en produkt. Eftersom tilldelningen inte är slumpmässig finns det ingen garanti för att grupperna kommer att vara jämförbara i alla avseenden, så observationsstudier kan inte fastställa kausalitet, bara association. Effekterna av behandlingen kan påverkas av faktorer som fick vissa människor att hamna i kontrollgruppen och vissa människor att hamna i behandlingsgruppen. Det finns dock sätt att kontrollera för förväxlingsfaktorer, vilket kan hjälpa till att stärka tillförlitligheten hos slutsatser om association.

<span style="color:blue">7. Longitudinella vs tvärsnittsstudier </span>

Den sista viktiga skillnaden att göra är mellan longitudinella och tvärsnittsstudier. I en longitudinell studie följs samma deltagare under en tidsperiod för att undersöka effekten av behandlingen på responsen. I en tvärsnittsstudie samlas data in från en enda punkt i tiden. Om du vill undersöka effekten av ålder på längd skulle en tvärsnittsstudie mäta längderna på personer i olika åldrar och jämföra dem. Resultaten kommer dock att påverkas av födelseår och livsstil eftersom det är möjligt att varje generation blir längre. I en longitudinell studie skulle samma personer ha sina längder registrerade vid olika tidpunkter i sina liv, så förväxlingen elimineras. Det är viktigt att notera att longitudinella studier är dyrare och tar längre tid att genomföra, medan tvärsnittsstudier är billigare, snabbare och mer bekväma.
