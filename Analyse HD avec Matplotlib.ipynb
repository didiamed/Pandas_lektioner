```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv("house_pricing.csv")
df.columns
df.head()
df.shape
```

```python
df.columns
df.head()
df.shape
```

```python
import numpy as np
import scipy.stats as stats

kde = stats.kde.gaussian_kde(df['SalePrice'])
x= np.linspace(0,750000, 100)
y=kde(x)

fig = plt.figure(figsize=(12,7))

plt.subplot(221)
plt.hist([df.SalePrice], color='b', label =['Prix de vente'] )
plt.title('distrubition')
plt.legend()

plt.subplot(222)
plt.plot(x, y, c = 'r', label = "Prix de vente")
plt.title('distrubition avec la m√©thode du noyau gaussien')
plt.legend()
plt.show()
```

```python

fig = plt.figure(figsize=(17,11))

plt.subplot(221)

plt.scatter(df['TotalBsmtSF'], [df.SalePrice], color='b', label ='Les prix de vente en fonction de  la surface de la cave' )
plt.legend()

plt.subplot(222)
plt.scatter(df['GrLivArea'], [df.SalePrice], color='y', label ='Les prix de vente en fonction de  la surface habitable' )
plt.legend()
plt.show()
```

```python

""" 
Il semble que les deux suivent la m√™me tendence. 
Les prix de vente augmente quand 
la surface habitable augmente et pareillement quand la surface de la cave monte. 

C'est tr√®s probablement d√ª au fait que plus une maison est grande plus
son prix monte et c'est dans les plus grandes maisons 
que l'on rencontre √† la fois des surfaces habitables plus grandes
et √©galement des surfaces de caves plus grandes.
"""
```
